## 2.1 Analysis of Literacy Rate vs Total Crimes.

```python
df1=df[['STATE/UT','YEAR','Total Crime',' LITERACY RATE %   (Male)',' LITERACY RATE % Female',' LITERACY RATE %  Average']]
df1
```

```python
df1.info()
```

```python
A1=df1.groupby('STATE/UT').mean().reset_index().sort_values(by=' LITERACY RATE %  Average',ascending=False)
A1
```

```python
plt.figure(figsize=(15,5))
plt.title('Average Literacy Rates V/S State/UTs')
plt.xticks(rotation=90)
plt.xlabel("STATES/UT")
plt.ylabel("AVERAGE LIteracy rate")
sns.barplot(x = 'STATE/UT', y = ' LITERACY RATE %  Average',data=df1,palette='Paired')
plt.show()
```

Observation: 
    1. 3 States i.e. Kerala, Mizoram, AND Lakshadweep have the highest average literacy rate.
    2. Bihar has the Lowest average Literacy Rate among all the states due to many reasons like Poverty, Lack of Knowledge in School Administration, Lack of Good Teachers in Schools, Poor Infrastruture, Child Labour and Lack of Awareness.



```python
plt.figure(figsize=(8,5))
sns.distplot(df1[' LITERACY RATE %   (Male)'], color='Green',label="Male",hist=False)
sns.distplot(df1[' LITERACY RATE % Female'],color='Blue',label="Female",hist=False)
plt.title("Distribution of Average Literacy Rate Between Males and Females")
plt.legend()
plt.show()
```

Observation: It has been clearly seen that Males have high literacy rates than the women.

In some parts of India, Child Marriage Pratha exists. Beacause of this they could not able to attend the school regularly or continue the eaducation and even parents do not send their girls childs in school because of safety issues, biasedness with sons, gender based violence , early marriage act and child labour in homes also.

```python
plt.figure(figsize=(8,5))
sns.regplot(x=df1[' LITERACY RATE %  Average'],y=df1['Total Crime'])
plt.xlabel(" LITERACY RATE %  Average")
plt.ylabel("Total Crime ")
plt.title("TOTAL CRIMES AGAINST CHILDREN V/S LITERACY RATE %  Average")
plt.grid()
plt.show()
```

It can be seen that there is no strong relationship between Literacy Rate % Average and Total Crime and there is a negative correlation between them 

```python
plt.figure(figsize=(8,5))
sns.regplot(x=df1[' LITERACY RATE %   (Male)'],y=df1['Total Crime'])
plt.xlabel(" LITERACY RATE %   (Male)")
plt.ylabel("Total Crime ")
plt.title("TOTAL CRIMES AGAINST CHILDREN V/S  LITERACY RATE %   (Male)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x=df1[' LITERACY RATE % Female'],y=df1['Total Crime'])
plt.xlabel(" LITERACY RATE % Female")
plt.ylabel("Total Crime ")
plt.title("TOTAL CRIMES AGAINST CHILDREN V/S LITERACY RATE % Female")
plt.grid()
plt.show()
```

```python
A1=df1.groupby('YEAR').sum().reset_index()[['YEAR','Total Crime']]
A1
plt.figure(figsize=(8,8))
sns.lineplot(x=df1['YEAR'],y=df1['Total Crime'])
plt.xticks([x for x in range(2001,2011)])
plt.xlabel("YEARS")
plt.ylabel("TOTAL CRIMES ")
plt.title("TOTAL CRIMES AGAINST CHILDREN OVER THE YEARS")
plt.grid()
plt.show()
```
